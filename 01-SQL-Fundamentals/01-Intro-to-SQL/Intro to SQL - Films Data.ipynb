```python
# Full dataset has 4968 rows

import pandas as pd

url = 'https://raw.githubusercontent.com/datacamp/courses-introduction-to-sql/master/data/films/films.csv'

df = pd.read_csv(url, index_col=0)

# reformat the release_year column
df['release_year']= df['release_year']/10
df['release_year']= (df['release_year']).astype(str).str.replace('.', '')

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df)
```
