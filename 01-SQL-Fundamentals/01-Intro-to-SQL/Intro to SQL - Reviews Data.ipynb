```python
# Full dataset has 4968 rows

import pandas as pd

url = 'https://raw.githubusercontent.com/datacamp/courses-introduction-to-sql/master/data/films/reviews.csv'

df = pd.read_csv(url, index_col=0)

# reformat the num_user column
df['num_user']= df['num_user']/10
df['num_user']= (df['num_user']).astype(str).str.replace('.', '')

# reformat the num_critic column
df['num_critic']= df['num_critic']/10
df['num_critic']= (df['num_critic']).astype(str).str.replace('.', '')

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df)
```
